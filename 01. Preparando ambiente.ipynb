```python
%sql
CREATE CATALOG IF NOT EXISTS bikestore
COMMENT 'Catalogo do projeto de bikes da InfinityBigData'
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS bikestore.resource
COMMENT 'Dados de origem e entrega';

CREATE SCHEMA IF NOT EXISTS bikestore.bronze
COMMENT 'Dados crus (raw)';

CREATE SCHEMA IF NOT EXISTS bikestore.silver
COMMENT 'Dados tratados';

CREATE SCHEMA IF NOT EXISTS bikestore.gold
COMMENT 'Dados agregados';

CREATE VOLUME IF NOT EXISTS bikestore.resource.origem
COMMENT 'Arquivos para clientes ou setores';

CREATE VOLUME IF NOT EXISTS bikestore.resource.destino
COMMENT 'Arquivos recebidos da fonte';
```
