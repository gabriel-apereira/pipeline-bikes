```python
from pyspark.sql import functions as F

bronze_path = 'bikestore.bronze'
silver_path = 'bikestore.silver'

# Carrega a tabela bronze
customers = spark.table(bronze_path+".ustomers")

# Aplica os filtros equivalentes ao WHERE
df_customer_silver = (
    customers
    .filter(F.col("phone").isNotNull())
    .filter(~F.col("phone").isin("NULL", "NULL "))
    .filter(F.col("email").isNotNull())
    .filter(~F.col("email").isin("NULL", "NULL "))
    .select(
        "customer_id",
        "first_name",
        "last_name",
        "phone",
        "email",
        "street",
        "city",
        "state",
        "zip_code"
    )
)

df_customer_silver.write.mode("overwrite").\
                    option('mergeSchema', 'true').\
                    saveAsTable(silver_path+".customers")
```
