```python
from pyspark.sql import functions as F

silver_path = 'bikestore.silver'
gold_path = 'bikestore.gold'

# Carrega a tabela silver
orders = spark.table(silver_path+".orders")

# Aplica filtros e agrega
df_sales_ny_gold = (
    orders
    .filter(F.col("state") == "NY")
    .filter(F.col("status") == "Delivered")
    .filter(F.col("shipped_date").isNotNull())
    .groupBy("shipped_date")
    .agg(F.round(F.sum("total_sale"), 2).alias("total_sale"))
)

df_sales_ny_gold.write.mode("overwrite").\
                    option('mergeSchema', 'true').\
                    saveAsTable(gold_path+".sales_ny")
```
